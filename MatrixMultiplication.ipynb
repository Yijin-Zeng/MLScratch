```python
def matrixmul(a,b):
    n_1 = len(a)
    m_1 = len(a[0])
    n_2 = len(b)
    m_2 = len(b[0])
    if m_1 != n_2:
        # shape does not match
        return -1
    output = [[0 for _ in range(m_2)] for _ in range(n_1)]
    for row in range(n_1):
        for col in range(m_2):
            for i in range(m_1):
                output[row][col] += (a[row][i] * b[i][col])
                
    return output
```

```python
A = [[1,2],[2,4]]
B = [[2,1],[3,4]]
```

```python
matrixmul(A,B)
```
