```python
from typing import Tuple
```

```python
def early_stopping(val_losses: list[float], patience: int, min_delta: float) -> Tuple[int, int]:
    best_val = float('inf')
    best_loc = -1

    for i, loss in enumerate(val_losses):
        if loss < best_val:
            best_val = loss
            best_loc = i

    stop_val = float('inf')
    patience_used = 0
    stop_loc = -1
    if_break = True
    for i, loss in enumerate(val_losses):
        if loss + min_delta < stop_val:
            stop_val = loss
            patience_used = 0
        else:
            patience_used += 1
        if patience_used == patience:
            stop_loc = i
            if_break = False
            break

    if if_break:
        stop_loc = len(val_losses) - 1

    return (stop_loc, best_loc)
```

```python
early_stopping([0.9, 0.8, 0.75, 0.77, 0.76, 0.77, 0.78], patience=2, min_delta=0.01)
```
