```python
import numpy as np
```

```python
def l1_regularization_gradient_descent(X: np.array, y: np.array, 
									   alpha: float = 0.1, learning_rate: float = 0.01, 
									   max_iter: int = 10000, tol: float = 1e-4) -> tuple:
	n, m = X.shape
	y = y.reshape(-1,1)
    # initialize weights and bias
	#weights = np.zeros(shape=(m,1)) # try different initialization
	weights = np.random.normal(size=(m,1))
	bias = 0
	iter_count = 0
	
	# iterate 
	update = 1
	while iter_count < max_iter and update > tol:
		# break conditions
		bias_col = np.ones(shape=(n,1)) * bias		
		# compute gradient
		X_beta = np.dot(X, weights)
		derivate_weights = - 1/n * np.dot((y - X_beta - bias_col).T, X) + alpha * np.sign(weights).T
		derivate_bias = - np.mean(y - X_beta - bias_col)

		# update weights and bias
		weights = weights - learning_rate*derivate_weights.T
		bias = bias - learning_rate*derivate_bias

		# update iter_count, and compute the l1 of derivate for break
		update = np.sum(np.abs(derivate_weights))
		iter_count += 1

	# Your code here
	return weights, bias, iter_count
```
