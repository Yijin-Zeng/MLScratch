```python
import numpy as np
```

```python
def transform_matrix(A, T, S):
	# Perform T^{-1} A S

	A = np.array(A)
	T = np.array(T)
	S = np.array(S)

	epsilon = 0.001
	if np.abs(np.linalg.det(T)) < epsilon or np.abs(np.linalg.det(S)) < epsilon:
		return -1
	else:
		transformed_matrix = np.linalg.inv(T) @ A @ S

	return transformed_matrix.tolist()
```

```python
A = [[1, 2], [3, 4]]
T = [[2, 0], [0, 2]]
S = [[1, 1], [0, 1]]
```

```python
transform_matrix(A, T, S)
```
