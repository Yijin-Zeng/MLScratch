```python
import numpy as np
```

```python
X = np.random.normal(0, 1, size=(105,))
y = np.random.uniform(0, 1, size=(105,))
```

```python
def k_fold_cross_validation(X: np.ndarray, y: np.ndarray, k=5, shuffle=True):
    n = len(X)
    if shuffle:
        random_index = np.random.permutation(range(n))
        X = X[random_index]
        y = y[random_index]
    
    batch_size = n//k
    validation_size = n - batch_size*(k-1)
    res = []
    for batch in range(k):
        validation_X = X[batch * batch_size: batch * batch_size + validation_size]
        validation_y = y[batch * batch_size: batch * batch_size + validation_size]
        training_X = np.concatenate((X[:batch * batch_size], X[batch * batch_size + validation_size:]), axis = 0)
        training_y = np.concatenate((y[:batch * batch_size], y[batch * batch_size + validation_size:]), axis = 0)
        res.append((training_X, training_y, validation_X, validation_y))
    return res
```

```python
res = k_fold_cross_validation(X,y)
```

```python
a,b,c,d = res[4]
```
