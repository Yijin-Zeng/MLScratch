```python
import numpy as np
```

```python
class NaiveBayes():
    def __init__(self, smoothing=1.0):
        # Initialize smoothing
        
        # empirical prior Prob(yi = 1)
        self.empirical_prior = 0
        
        #### improvement: can use a diction to store dict[0] = self.likelihood_y_0
        
        # likelihood function Prob(x = 1|y=1)
        self.likelihood_y_1 = None  
        # likelihood function Prob(x = 1|y=0)
        self.likelihood_y_0 = None
        
        # store the dimension
        self.d = 0
        # smoothing
        self.smoothing = smoothing

    def forward(self, X, y):
        # Fit model to binary features X and labels y
        n, self.d = X.shape
        k = np.sum(y == 1) 
        self.empirical_prior = k / n
        self.likelihood_y_1 = (np.sum(X[y == 1], axis = 0) + self.smoothing)/(2*self.smoothing + k) # apply Laplacian smoothing
        self.likelihood_y_0 = (np.sum(X[y == 0], axis = 0) + self.smoothing)/(2*self.smoothing + n - k) # apply Laplacian smoothing

    def predict(self, X):
        output = []
        X = X.reshape(-1,self.d)
        n = X.shape[0]
        for k in range(n):
            sample = X[k]
            # Predict class labels for test set X
            cond_likelihood_y_1 = 1
            for i in range(self.d):
                if sample[i] == 1:
                    cond_likelihood_y_1 *= self.likelihood_y_1[i]
                else:
                    cond_likelihood_y_1 *= 1 - self.likelihood_y_1[i]

            cond_likelihood_y_0 = 1
            for i in range(self.d):
                if sample[i] == 1:
                    cond_likelihood_y_0 *= self.likelihood_y_0[i]
                else:
                    cond_likelihood_y_0 *= 1 - self.likelihood_y_0[i]

            # compute the posterior probability
            post = self.empirical_prior * cond_likelihood_y_1 / (self.empirical_prior * cond_likelihood_y_1 + (1 - self.empirical_prior) * cond_likelihood_y_0)

            output.append(post)

        return output
```

```python
model = NaiveBayes()
```

```python
X = np.array([[0, 0], [1, 0], [0, 1]]) 
y = np.array([0, 1, 0]) 
model.forward(X, y)
```
