```python
import numpy as np
```

```python
def Kmeans(data, initial_centroids, max_iter):
    n = data.shape[0]
    k = initial_centroids.shape[0]
    count = 0
    centroids = initial_centroids
    while count < max_iter:
        # assign clusters to each instance
        dis = np.zeros(shape=(n,k))
        for i in range(n):
            for j in range(k):
                dis[i][j] = np.sum(np.square(data[i,:] - centroids[j,:]))

        clusters = np.argmin(dis, axis=1)
        # update centroids
        for i in range(k):
            centroids[i] = np.mean(data[clusters == i], axis=0)
        count += 1
    return centroids
```

```python
def KmeansPlusPlus(data, k):
    # automatic initialize centroids
    # note: the initialized centroids are instances
    n = data.shape[0]
    initial_centroids = []
    p = np.ones(shape=(n,))*(1/n) #initilize probability
    dis = np.zeros(shape=(k, n)) # here we store the distance of instances to each centroid
    for i in range(k):
        index = np.random.choice(n, size=1, p=p)   # sample 1 row index
        initial_centroids.append(data[index[0]])  # get the actual row from data
        for j in range(n):
            dis[i,j] = np.sum(np.square(data[j,:] -  initial_centroids[-1]))
        # update probability
        p = np.square(np.min(dis[:i+1,:], axis = 0)) / np.sum(np.square(np.min(dis[:i+1,:], axis = 0)))

    return initial_centroids

```
