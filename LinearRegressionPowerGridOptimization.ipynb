```python
import math
import numpy as np
PI = 3.14159
```

```python
def power_grid_forecast(consumption_data):
	# 1) Subtract the daily fluctuation (10 * sin(2π * i / 10)) from each data point.
	# 2) Perform linear regression on the detrended data.
	# 3) Predict day 15's base consumption.
	# 4) Add the day 15 fluctuation back.
	# 5) Round, then add a 5% safety margin (rounded up).
	# 6) Return the final integer.
    
    ### start
    # 1
    consumption_data = np.array(consumption_data)
    consumption_data = consumption_data -  10 * np.sin(2*PI*np.arange(1,11) / 10)


    # 2 
    X = np.stack((np.ones_like(consumption_data), np.arange(1,11)), axis=1)
    beta = np.linalg.inv(X.T @ X) @ X.T @ consumption_data
    
	# 3
    prediction = beta @ np.array([1,15])
    
	# 4
    prediction = prediction + 10 * np.sin(2*PI*15 / 10)
    
	# 5
    res = math.ceil(prediction * 1.05)
    
	# 6
    return res
    
```

```python
power_grid_forecast(consumption_data)
```
