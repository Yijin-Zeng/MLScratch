```python
import numpy as np
```

```python
def GramSchmidt(A):
    # A: a matrix with independent columns
    n, m = A.shape
    Q = np.zeros_like(A)
    R = np.zeros(shape=(m,m))
    for i in range(m):
        a_i = A[:,i]
        b_i = np.copy(a_i)
        for j in range(i):
            R[j,i] = np.dot(Q[:,j], a_i)
            b_i -= R[j,i] * Q[:,j]

        Q[:,i] = b_i
        # standarize the column
        scale = np.sqrt(np.dot(Q[:,i], Q[:,i]))
        Q[:,i] = Q[:,i] / scale
        R[i,i] = scale
    return Q, R
```

```python
A = np.random.normal(0,1,size=(5,3))
```

```python
A
```

```python
Q, R = GramSchmidt(A)
```

```python
tol = 0.00001
np.sum(np.square(np.matmul(Q,R) - A)) < tol
```
