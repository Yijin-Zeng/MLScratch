```python
import numpy as np
```

```python
def gaussian_elimination(A, b):
	"""
	Solves the system Ax = b using Gaussian Elimination
	:param A: Coefficient matrix
	:param b: Right-hand side vector
	:return: Solution vector x
	"""
	n = len(A)
	for i in range(n-1):
		# first find pivot
		k = i + np.argmax(np.abs(A[i:,i]))
		if k != i:
			# swap ith row and kth row
			A[[i, k]] = A[[k, i]]
			b[[i, k]] = b[[k, i]]
		pivot = A[i,i]
		
		# forward elimination
		for j in range(i+1,n):
			l = A[j,i] / pivot
			# update A
			A[j,:] = A[j,:] - A[i,:] * l
			# update b
			b[j] = b[j] - b[i] * l
	
	# back substitution
	output = np.array([0 for _ in range(n)], dtype=float) # dont forget to assign dtype
	# even better
    # output = np.zeros_like(b)
	output[-1] = b[-1] / A[-1,-1]
	for i in range(n-2,-1,-1):
		output[i] = (b[i] - np.dot(A[i,i+1:], output[i+1:])) / (A[i,i])

	return output[:n]
```

```python
A = np.array([ [0, 2, 1, 0, 0, 0, 0], [2, 6, 2, 1, 0, 0, 0], [1, 2, 7, 2, 1, 0, 0], [0, 1, 2, 8, 2, 1, 0], [0, 0, 1, 2, 9, 2, 1], [0, 0, 0, 1, 2, 10, 2], [0, 0, 0, 0, 1, 2, 11] ], dtype=float)
b = np.array([1, 2, 3, 4, 5, 6, 7], dtype=float)
gaussian_elimination(A,b)
```
