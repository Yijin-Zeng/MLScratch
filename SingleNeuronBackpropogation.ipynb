```python
import numpy as np
```

```python
def train_neuron(features: np.ndarray, labels: np.ndarray, initial_weights: np.ndarray, initial_bias: float, learning_rate: float, epochs: int) -> (np.ndarray, float, list[float]):
	# Your code here
	# first make sure the shape for all inputs are correct
	labels = labels.reshape(-1,1)
	initial_weights = initial_weights.reshape(-1,1)
	initial_bias = np.array(initial_bias).reshape(-1,1)
	N = len(labels)

	def forward_pass(inputs, weights, bias):
		t = 1 + np.exp(-(inputs @ weights + bias))
		pred = 1 / t
		return pred

	def get_gradient(pred, inputs, labels, weights, bias):
		z = pred - labels
		alpha_l_alpha_z = (2 / N) * z.T
		alpha_z_alpha_t = np.diag((-pred**2).flatten())
		alpha_t_alpha_w = -(1/pred-1) * inputs
		gradients_weights = alpha_l_alpha_z @ alpha_z_alpha_t @ alpha_t_alpha_w
		gradient_bias = alpha_l_alpha_z @ alpha_z_alpha_t @ -(1/pred-1)
		return gradients_weights.T, gradient_bias


	mse_values = []
	weights = initial_weights
	bias = initial_bias
	inputs = features
	for _ in range(epochs):
		
		predictions = forward_pass(inputs, weights, bias)

		mse_values.append(np.round(np.mean(np.square(predictions - labels)),4))

		gradients_weights, gradient_bias = get_gradient(predictions, inputs, labels, weights, bias)

		weights -= learning_rate * gradients_weights

		bias -= learning_rate * gradient_bias


	return np.round(weights, 4), np.round(bias,4), mse_values
```

```python
features = np.array([[1.0, 2.0], [2.0, 1.0], [-1.0, -2.0]])
labels = np.array([1, 0, 0])
initial_weights = np.array([0.1, -0.2])
initial_bias = 0.0
learning_rate = 0.1
epochs = 2
```

```python
train_neuron(features, labels, initial_weights, initial_bias, learning_rate, epochs)
```
