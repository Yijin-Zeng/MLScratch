```python
import numpy as np
```

```python
def calculate_eigenvalues(matrix):
    # compute trace and determinant
    trace = matrix[0][0] + matrix[1][1]
    determinant = matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0]

    # note: lambda_1 + lambda_2 = trace
    # and lambda_1 * lambda = determinant

    # (lambda_1 - lambda_2)^2
    square_difference = trace**2 - 4 * determinant

    lambda_1 = (trace + np.sqrt(square_difference))/2
    lambda_2 = (trace - np.sqrt(square_difference))/2

    return lambda_1, lambda_2
```

```python
matrix = [[2, 1], [1, 2]]
```

```python
calculate_eigenvalues(matrix)
```
